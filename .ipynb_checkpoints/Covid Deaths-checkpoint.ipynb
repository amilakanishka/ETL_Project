# Count covid deaths by country and put in output file

###### import library

```python
import pandas as pd
```

###### read data from main dataset

```python
death_file = "Resources/covid_19_deaths.csv"
death_df = pd.read_csv(death_file, encoding="utf-8")
death_df
```

###### change column names to make them more meaningful 

```python
death_df = death_df.rename(columns={"Country/Region":"Country", "8/12/2020":"Covid Deaths till 12th August, 2020"})
death_df
```

###### call only necessary columns

```python
death_df = death_df[["Country","Covid Deaths till 12th August, 2020"]]
death_df
```

###### groupby countries to avoid multiple entries of the country name and sum the values

```python
groupby_country = death_df.groupby(['Country']).sum()
groupby_country
```

###### create an output file to upload in database schema

```python
groupby_country.to_csv("Covid deaths by country.csv", index=True, header=True)
```
