# Count covid deaths by country and put in output file

###### import library

```python
import pandas as pd
```

###### read data from main dataset

```python
death_file = "Resources/covid_19_deaths.csv"
death_df = pd.read_csv(death_file, encoding="utf-8")
death_df.head()
```

###### change column names to make them more meaningful 

```python
death_df = death_df.rename(columns={"Country/Region":"country", "8/12/2020":"death"})
death_df.head()
```

###### call only necessary columns

```python
death_df = death_df[["country","death"]]
death_df.head()
```

```python
covid_file = "Resources/covid_19_confirmed.csv"
confirmdf = pd.read_csv(covid_file, encoding="utf-8")
confirmdf.head()
```

```python

confirmdf = confirmdf.rename(columns={"Country/Region":"country", "8/12/20":"confirmed"})
confirmdf = confirmdf[["country","confirmed"]]
confirmdf.head()
```

```python
coviddf = death_df.merge(confirmdf, left_on='country', right_on='country')
coviddf
```

###### create an output file to upload in database schema

```python
coviddf.to_csv("Covid_stats.csv", index=False, header=True)
```
