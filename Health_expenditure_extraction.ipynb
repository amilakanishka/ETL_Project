```python
import pandas as pd
```

```python
path = "Resources/raw_health_expenditure.csv"
health_df = pd.read_csv(path, encoding="utf-8")
health_df.head()
```

```python
health_df = health_df[["Country", "Value"]]

health1 = health_df.sort_values('Country')
health1.head()
```

```python
health = health1.rename(columns={"Country": "country", "Value": "health_expenditure"})
health
```

```python
health.to_csv("Data/health_expenditure_by_Country.csv", index=False, header=True)
```
