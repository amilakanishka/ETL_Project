```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
death_file = "Resources/covid_19_deaths.csv"
death_df = pd.read_csv(death_file, encoding="utf-8")
death_df
```

```python
death_df = death_df.rename(columns={"Country/Region":"Country", "Total":"Covid Deaths"})
death_df
```

```python
death_df = death_df[["Country","Covid Deaths"]]
death_df
```

```python
groupby_country = death_df.groupby(['Country']).sum()
groupby_country
```

```python
groupby_country.to_csv("Covid deaths by country.csv", index=True, header=True)
```
