```python
import pandas as pd
import numpy as np
```

```python
master = pd.read_csv("Master Countries.csv") 
master
```

```python
master= master[["Country (OECD)","Region"]]
master
```

```python
#transforming the data into a cleaner format by dropping and renaming columns
category = pd.read_csv("country_category.csv")
category= category.drop(['Country Code', 'SpecialNotes'], axis=1)
category.head()
```

```python
category['TableName'] = category['TableName'].replace(['China'],"China (People's Republic of)")
category['TableName'] = category['TableName'].replace(['Korea, Rep.'],"Korea")
category['TableName'] = category['TableName'].replace(['Russian Federation'],"Russia")
```

```python
#transforming the data into a cleaner format by dropping and renaming columnscategory
category= category.rename(columns={'TableName': 'country', 'IncomeGroup': 'income_group'})
category.head()
```

```python
#joining the tables
final_table = pd.merge(left=master, right=category, how='left', left_on='Country (OECD)', right_on='country')
final_table= final_table.drop(['country'], axis=1)
```

```python
final_table= final_table.rename(columns={'Region_x': 'EU', 'Region_y': 'Region'})
```
